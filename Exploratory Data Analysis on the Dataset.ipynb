```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load dataset
file_path = "C:/Users/User/Final Year Project/cleaned_dataset.csv"
data = pd.read_csv(file_path)
```

```python
# Descriptive statistics
print(data.describe())
```

```python
# Visualize inventory trends
plt.figure(figsize=(10, 6))
sns.lineplot(data=data, x='ExpDate', y='Quantity', label='Stock Levels')
plt.title('Stock Level Trends Over Time')
plt.xlabel('ExpDate')
plt.ylabel('Quantity')
plt.legend()
plt.show()
```

```python
# Analyze fast-moving products
fast_moving = data.groupby('DrugName')['Price'].sum().sort_values(ascending=False).head(10)
fast_moving.plot(kind='bar', figsize=(10, 6), title='Top 10 Fast-Moving Products')
plt.xlabel('DrugName')
plt.ylabel('Price')
plt.show()
```
