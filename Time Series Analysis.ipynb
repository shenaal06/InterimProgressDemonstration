```python
from statsmodels.tsa.arima.model import ARIMA
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Load dataset
file_path = "C:/Users/User/Final Year Project/cleaned_dataset.csv"
data = pd.read_csv(file_path)
```

```python
# Aggregate data by date
data['ExpDate'] = pd.to_datetime(data['ExpDate'])
time_series = data.groupby('ExpDate')['Quantity'].sum()
```

```python
# Fit ARIMA model
model = ARIMA(time_series, order=(1, 1, 1))
results = model.fit()
```

```python
# Forecast
forecast = results.forecast(steps=30)
plt.figure(figsize=(10, 6))
plt.plot(time_series, label='Historical Sales')
plt.plot(forecast, label='Forecasted Sales', linestyle='--')
plt.legend()
plt.show()
```
